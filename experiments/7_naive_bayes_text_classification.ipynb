# Experiment 7: Naïve Bayesian Classifier for Document Classification

## Overview
This experiment demonstrates text classification using Naïve Bayes. A common application is spam detection, sentiment analysis, or document categorization. We'll use a sample dataset and compute accuracy, precision, and recall metrics.

## Key Concepts
- **Vectorization**: Converting text to numerical features (TF-IDF or Bag of Words)
- **Feature Extraction**: Selecting important words/tokens from documents
- **Naïve Bayes**: Probabilistic classifier for text data
- **Performance Metrics**: Accuracy, Precision, Recall, F1-Score

````python
# Naive Bayes for Document Classification

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Custom documents
documents = [
    "Team won the cricket match",
    "Player scored a century",
    "The batsman hit a six",
    "Football team played well",
    "NASA launched a satellite",
    "Space mission was successful",
    "Astronauts returned safely",
    "Rocket reached the moon"
]

# Labels
# Sports = 0, Space = 1
labels = [0, 0, 0, 0, 1, 1, 1, 1]

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    documents, labels, test_size=0.25, random_state=42
)

# Convert text into numbers
vectorizer = CountVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Create model
model = MultinomialNB()

# Train model
model.fit(X_train_vec, y_train)

# Predict
y_pred = model.predict(X_test_vec)

# Results
print("Predicted:", y_pred)
print("Actual   :", y_test)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
````

### Explanation

* `CountVectorizer()` converts documents into numeric word counts.
* `MultinomialNB()` is used for text/document classification.
* `accuracy_score()` checks overall correctness.
* `precision_score()` measures correct positive predictions.
* `recall_score()` measures how many actual positives were found.

## Viva Questions

### Basic Concepts
1. **Why is Naïve Bayes suitable for text classification?**
   - Works well with high-dimensional sparse data (text documents)
   - Fast training and prediction
   - Performs surprisingly well despite naive independence assumption
   - Interpretable results

2. **What is TF-IDF vectorization?**
   - Term Frequency-Inverse Document Frequency
   - Converts text to numerical features
   - TF: frequency of term in document
   - IDF: inverse frequency of term in corpus

3. **How does TF-IDF differ from Bag of Words?**
   - BoW: simple word count (no weighting)
   - TF-IDF: weights terms by importance
   - TF-IDF reduces impact of common words
   - TF-IDF generally performs better

### Implementation Questions
4. **What preprocessing steps are important for text?**
   - Lowercasing: standardize case
   - Removing stop words: eliminate common non-informative words
   - Tokenization: split text into words
   - Stemming/Lemmatization: normalize word forms

5. **Why use stop word removal?**
   - Common words (the, is, a) add noise
   - Reduces feature dimensionality
   - Improves classifier performance
   - Faster processing

6. **How does MultinomialNB differ from GaussianNB?**
   - MultinomialNB: for discrete count data (text, word frequencies)
   - GaussianNB: for continuous data with normal distribution
   - MultinomialNB commonly used for text classification

### Evaluation Questions
7. **What are common evaluation metrics for classification?**
   - Accuracy: overall correctness
   - Precision: true positives / predicted positives
   - Recall: true positives / actual positives
   - F1-Score: harmonic mean of precision and recall

8. **When should you use precision vs recall?**
   - Precision: when false positives are costly (spam detection)
   - Recall: when false negatives are costly (disease detection)
   - F1-Score: when balance between both is important

9. **What does the confusion matrix show?**
   - True Positives: correct positive predictions
   - True Negatives: correct negative predictions
   - False Positives: incorrect positive predictions
   - False Negatives: incorrect negative predictions

10. **How do you handle imbalanced text datasets?**
    - Adjust class weights in classifier
    - Use stratified train-test split
    - Oversample minority class
    - Undersample majority class
    - Use appropriate evaluation metrics
