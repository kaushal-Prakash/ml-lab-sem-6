### 1. Adaline

```python
import numpy as np

class Adaline:
    def __init__(self, learning_rate=0.01, n_iterations=50):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.cost = []
    
    def fit(self, X, y):
        self.weights = [0.0] * X.shape[1]
        self.bias = 0.0
        
        for _ in range(self.n_iterations):
            output = X.dot(self.weights) + self.bias
            errors = y - output
            self.weights += self.learning_rate * X.T.dot(errors)
            self.bias += self.learning_rate * errors.sum()
            cost = (errors ** 2).sum() / 2.0
            self.cost.append(cost)
            
    def predict(self, X):
        return np.where(X.dot(self.weights) + self.bias >= 0.5, 1, 0)
    
model = Adaline()
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
model.fit(X, y)
predictions = model.predict(X)
print(predictions)
```
